```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
from matplotlib.ticker import FormatStrFormatter
import itertools
warnings.filterwarnings('ignore')
from matplotlib.backends.backend_pdf import PdfPages
```

```python
import sys
sys.path.append('..')
```

```python

category_map = {'harm':'Care/Harm', 'fairness': 'Fairness/Cheating', 
               'authority': 'Authority/Subversion', 'loyalty': 'Loyalty/Betrayal',
               'purity':'Purity/Degradation'}
```

```python
v = 'weighted'
version = 300
at1 = ''
certain = '_certain'
f_vote = True
frequency_df = pd.read_csv(f'../../data/Survey Data/N300/freqs/v2_new_frequency_polarity_count_{v}_full_CHILDES_version_N{version}{at1}{certain}_fvote_{f_vote}_all.csv')
                    
gender_df = pd.read_csv(f'../../data/Survey Data/N300/freqs/v3_new_frequency_gender_count_{v}_full_CHILDES_version_N{version}{at1}{certain}_fvote_{f_vote}_all.csv')

```

```python
child_df = frequency_df.loc[frequency_df.identity == 'child']
parent_df = frequency_df.loc[frequency_df.identity == 'parent']
```

```python
def plot_combo_stack_area(frequency_df, gender_df, vote,at1, certain, f_vote, version = 300, savefig = True):
    pp = PdfPages('../../multiple.pdf')
    child_df = frequency_df.loc[frequency_df.identity == 'child']
    parent_df = frequency_df.loc[frequency_df.identity == 'parent']

    fig, axes = plt.subplots(1, 2, figsize = (15,4), sharey = False, sharex = False)
    ax = axes[0]


    full_labels = ['Care','Harm', 'Fairness', 'Cheating', 'Authority', 'Subversion', 'Loyalty', 'Betrayal', 'Purity','Degradation']
    d = parent_df[['moral foundation','age','moral frequency']]
    d = d.pivot(index='age',columns='moral foundation',values='moral frequency')
    d = d.reindex(columns=['harm pos', 'harm neg', 'fairness pos', 'fairness neg', 
                           'authority pos','authority neg', 'loyalty pos', 'loyalty neg', 'purity pos', 'purity neg'])



    d.plot.area(ax = ax)




    ax.set_xlabel('age', fontsize=12)
    ax.set_ylabel('percentage of moral language', fontsize=12)
    ax.set_title('Child-directed speech (CDS)',fontsize=14)
    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)

    handles, labels = ax.get_legend_handles_labels()
    _, handles = zip(*sorted(zip(labels, handles), key=lambda t: 1, reverse=True))
    
    ax.legend_ = None

    full_labels = ['Care','Harm', 'Fairness', 'Cheating', 'Authority', 'Subversion', 'Loyalty', 'Betrayal', 'Purity','Degradation']
    d = child_df[['moral foundation','age','moral frequency']]
    d = d.pivot(index='age',columns='moral foundation',values='moral frequency')
    d = d.reindex(columns=['harm pos', 'harm neg', 'fairness pos', 'fairness neg', 
                           'authority pos','authority neg', 'loyalty pos', 'loyalty neg', 'purity pos', 'purity neg'])


    ax = axes[1]

    d.plot.area(ax = ax)

    ax.set_xlabel('age', fontsize=12)

    ax.set_title('Child speech (CS)',fontsize=14)

    handles, labels = ax.get_legend_handles_labels()
    _, handles = zip(*sorted(zip(labels, handles), key=lambda t: 1, reverse=True))

    ax.legend(handles, full_labels, bbox_to_anchor = (1, 1))
    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)
    if savefig:
    
        plt.savefig(f'../../combined_{vote}_vote_moral_emergence_N{version}{at1}{certain}_fvote_{f_vote}_shrunk_1.png', bbox_inches = 'tight', dpi = 600)
    
    fig, axes = plt.subplots(1, 4, figsize = (15,2), sharey = True, sharex = True)
    #---------
    categories = ['fairness', 'purity']
    mapping = {'harm': 'Care/Harm', 'fairness':'Fairness/Cheating', 'authority': 'Authority/Subversion', \
               'loyalty': 'Loyalty/Betrayal','purity':'Purity/Degradation'}
    color_map = {0: ('tab:green', 'tab:red'), 1: ('tab:olive', 'tab:cyan'), }
    count = 0
    for i , c in enumerate(categories):
        cs_df = child_df.loc[child_df.category == c]
        cds_df = parent_df.loc[parent_df.category == c]

        d = cds_df[['sentiment','age','moral frequency']]
        d = d.pivot(index='age',columns='sentiment',values='moral frequency')
        d = d.reindex(columns=['pos', 'neg'])

        ax = axes[count]
#         ax = fig.add_subplot(312)
        count += 1
        d.plot.area(ax = ax,color=color_map[i])

        plt.xlabel('age', fontsize=12)
        ax.set_ylabel(f'', fontsize=12)

        ax.spines.right.set_visible(False)
        ax.spines.top.set_visible(False)
        ax.set_title(f'CDS ({mapping[c]})')

        handles, labels = ax.get_legend_handles_labels()
        _, handles = zip(*sorted(zip(labels, handles), key=lambda t: 1, reverse=True))
        ax.legend_ = None

        #for CS

        d = cs_df[['sentiment','age','moral frequency']]
        d = d.pivot(index='age',columns='sentiment',values='moral frequency')
        d = d.reindex(columns=['pos', 'neg'])

        ax = axes[count]
#         ax = fig.add_subplot(322)
        count += 1
        d.plot.area(ax = ax, color = color_map[i])
        plt.xlabel('age', fontsize=10)
        ax.set_ylabel(f'', fontsize=12)
        ax.set_title('CS',fontsize=14)


        
        ax.spines.right.set_visible(False)
        ax.spines.top.set_visible(False)
        handles, labels = ax.get_legend_handles_labels()
        _, handles = zip(*sorted(zip(labels, handles), key=lambda t: 1, reverse=True))

        slash_index = mapping[c].index('/')
        pos = mapping[c][:slash_index]
        neg = mapping[c][slash_index + 1: ]     
        ax.set_title(f'CS ({mapping[c]})')
        ax.legend_ = None
        ax.yaxis.set_major_formatter(FormatStrFormatter('%.3f'))



    


    
    
    
    
  

    if savefig:
    
        plt.savefig(f'../../combined_{vote}_vote_moral_emergence_N{version}{at1}{certain}_fvote_{f_vote}_shrunk_2.png', bbox_inches = 'tight', dpi = 600)
    
    #-----------------------------
    child_df = gender_df.loc[gender_df.identity == 'child']
    parent_df = gender_df.loc[gender_df.identity == 'parent']



    fig, axes = plt.subplots(1, 4, figsize = (15,2), sharey = True, sharex = True)
    full_labels = ['Care','Harm', 'Fairness', 'Cheating', 'Authority', 'Subversion', 'Loyalty', 'Betrayal', 'Purity','Degradation']
    d = parent_df.loc[parent_df.gender == 'female'][['moral foundation','age','moral frequency']]
    d = d.pivot(index='age',columns='moral foundation',values='moral frequency')
    d = d.reindex(columns=['harm pos', 'harm neg', 'fairness pos', 'fairness neg', 
                           'authority pos','authority neg', 'loyalty pos', 'loyalty neg', 'purity pos', 'purity neg'])
    
    ax = axes[0]
    d.plot.area(ax = ax)

    plt.xlabel('', fontsize=12)
    ax.set_ylabel('', fontsize=12)
    ax.set_title('CDS to female children',fontsize=14)
    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)
    # ax = plt.gca()

    handles, labels = ax.get_legend_handles_labels()
    _, handles = zip(*sorted(zip(labels, handles), key=lambda t: 1, reverse=True))
    ax.legend(handles, full_labels)

    ax.set_xlabel('', fontsize=10)
    ax.legend_ = None
    ax = axes[1]
    d = parent_df.loc[parent_df.gender == 'male'][['moral foundation','age','moral frequency']]
    d = d.pivot(index='age',columns='moral foundation',values='moral frequency')
    d = d.reindex(columns=['harm pos', 'harm neg', 'fairness pos', 'fairness neg', 
                           'authority pos','authority neg', 'loyalty pos', 'loyalty neg', 'purity pos', 'purity neg'])

    d.plot.area(ax = ax)

    plt.xlabel('', fontsize=12)
    ax.set_xlabel('', fontsize=10)
    ax.set_ylabel('', fontsize=12)
    ax.set_title('CDS to male children',fontsize=14)

    # ax = plt.gca()

    handles, labels = ax.get_legend_handles_labels()
    _, handles = zip(*sorted(zip(labels, handles), key=lambda t: 1, reverse=True))
    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)
    ax.legend_ = None
    ax = axes[2]
    d = child_df.loc[child_df.gender == 'female'][['moral foundation','age','moral frequency']]
    d = d.pivot(index='age',columns='moral foundation',values='moral frequency')
    d = d.reindex(columns=['harm pos', 'harm neg', 'fairness pos', 'fairness neg', 
                           'authority pos','authority neg', 'loyalty pos', 'loyalty neg', 'purity pos', 'purity neg'])

    d.plot.area(ax = ax)
    

    plt.xlabel('', fontsize=12)
    ax.set_xlabel('', fontsize=10)
    ax.set_title('CS of female children',fontsize=14)
    # 
    # ax = plt.gca()

    handles, labels = ax.get_legend_handles_labels()
    _, handles = zip(*sorted(zip(labels, handles), key=lambda t: 1, reverse=True))

    ax.legend(handles, full_labels)
    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)

#     ax.legend(handles, full_labels,bbox_to_anchor=(1, 1))
    

    ax.legend_ = None
    ax = axes[3]
    d = child_df.loc[child_df.gender == 'male'][['moral foundation','age','moral frequency']]
    d = d.pivot(index='age',columns='moral foundation',values='moral frequency')
    d = d.reindex(columns=['harm pos', 'harm neg', 'fairness pos', 'fairness neg', 
                           'authority pos','authority neg', 'loyalty pos', 'loyalty neg', 'purity pos', 'purity neg'])

    d.plot.area(ax = ax)

    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)

    ax.set_xlabel('', fontsize=10)
    ax.set_title('CS of male children',fontsize=14)

    # ax = plt.gca()
    handles, labels = ax.get_legend_handles_labels()
    _, handles = zip(*sorted(zip(labels, handles), key=lambda t: 1, reverse=True))

    ax.legend_ = None


    if savefig:
    
        plt.savefig(f'../../combined_{vote}_vote_moral_emergence_N{version}{at1}{certain}_fvote_{f_vote}_shrunk_3.png', bbox_inches = 'tight', dpi = 600)
    
    
    plt.show()
    

```

```python
plot_combo_stack_area(frequency_df, gender_df, vote=v, f_vote=f_vote, at1 = at1, 
                      certain = certain, version = version, savefig=True)



```
